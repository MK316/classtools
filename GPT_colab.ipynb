```python
#@markdown Paste your API key
import openai
from gtts import gTTS
from IPython.display import Audio

# Set up OpenAI API credentials
mykey = input("Paste your key: ")
openai.api_key = mykey
```

```python
#@markdown Q & As (Ask a question and the response will be converted to speech and played using gTTS.)

# Function to generate response from OpenAI chatbot
def generate_response(prompt):
    response = openai.Completion.create(
        engine="davinci",
        prompt=prompt,
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.5,
    )
    return response.choices[0].text.strip()
# Take user input and generate chatbot response
user_input = input("Ask a question: ")
response = generate_response(user_input)

print(response)

# revision = input("Do you want me to revise the answer? (y or n)")

# if revision == "y":
#   user_input2 = input("How may I revise it?")
#   response2 = generate_response(user_input2)
#   final_response = response2
#   print("Revised answer: %s"%response2)

# elif revision == "n":
#   print("Okay.")
#   print("*"*50)

# continue_button = input("Generate speech? (y or n) ")

# if continue_button == "y":
tts = gTTS(response)
tts.save("response.mp3")

# elif continue_button == "n":
#     tts = gTTS("Okay. I'll be quiet. ")
#     tts.save("response.mp3")
# Play the audio file automatically
Audio("response.mp3", autoplay=True)
```
